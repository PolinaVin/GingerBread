```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
def func(x):
    return (2+np.cos(1*x**0.5))*(1+0.5*x)**(-0.5)*np.exp(x/2)
```

```python
start=0
end=2
x=np.linspace(start, end, 10*6)
weight=x[1]-x[0]
f=func(x)
F=list()
f1=list()
I=0
for xi in x:
    I+=weight*func(xi)
    F.append(I)
    if len(F)>1:
        f1.append((F[-1]- F[-2])/weight)
print(I)
plt.plot(x, f, '-', x, F, '-', x[1:], f1, '-')
```
